```python
from copy import deepcopy
from itertools import product
from pprint import pprint

from docarray import DocumentArray, Document

from weave_search.semantic_search.data_utils import get_information_for_term
```

```python
TERMS = [
    "apoptosis", "lung cancer",  "alzheimer", 
]
DATA_CATEGORIES = [
    "Protein", "Pathway", "Reaction", "Complex"
]
SUBCATEGORIES = [
    "pathologies", "nonpathologies"
]
```

```python
all_data = {term: get_information_for_term(term) for term in TERMS}
```

```python
document_keys = set()
i = 0

# Get all document keys
for term, category, sub_category in product(*[TERMS, DATA_CATEGORIES, SUBCATEGORIES], repeat=1):
    doc = all_data[term][category][sub_category]
    
    doc_list = deepcopy(doc)
    if not isinstance(doc_list, list):
        doc_list = doc.values()
    
    for item in doc:
        if "summation" in item:
            i += 1
            document_keys.update(list(item.keys()))

print(f"{i} documents found")
print(document_keys)
```

```python
# Print one document from each term
for term, category, sub_category in product(*[TERMS, DATA_CATEGORIES, SUBCATEGORIES], repeat=1):
    doc = all_data[term][category][sub_category]
    
    doc_list = deepcopy(doc)
    if not isinstance(doc_list, list):
        doc_list = doc.values()
    
    for item in doc:
        if "summation" in item:
            pprint(item)
            break
```
